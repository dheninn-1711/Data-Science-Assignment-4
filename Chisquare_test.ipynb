```python
#import libraries
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
```

```python
#data observed
data = np.array([[50, 70], [80, 100], [60, 90], [30, 50], [20, 50]])
data
```

```python
#Create a DataFrame
df = pd.DataFrame(data, columns=['Smart_Thermostat', 'Smart_Light'],
                  index=['Very Satisfied', 'Satisfied', 'Neutral', 'Unsatisfied', 'Very Unsatisfied'])
df
```

```python
#Stating Hypothesis
'''Null Hypothesis (H0): There is no significant association between the type of smart home device purchased and the customer satisfaction level.
   Alternative Hypothesis (H1): There is a significant association between the type of smart home device purchased and the customer satisfaction level.'''
```

```python
#Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(df)
```

```python
#Print results
print(f"Chi-Square Statistic: {chi2}")#test statistic computed from the observed and expected frequencies
print(f"P-Value: {p}")#the probability of observing the data if the null hypothesis is true
print(f"Degrees of Freedom: {dof}")#number of categories
print("The Expected Frequencies are:")
#Expected frequencies are shown
pd.DataFrame(expected, columns=['Smart_Thermostat', 'Smart_Light'],
                   index=['Very Satisfied', 'Satisfied', 'Neutral', 'Unsatisfied', 'Very Unsatisfied'])
```

```python
#Determine the critical value
alpha = 0.05
critical_value = chi2_contingency(df)[0]#cutoff point beyond which we reject the null hypothesis
print(f"\nCritical Value at alpha={alpha}: {critical_value}")
```

```python
#Make a decision
if chi2 > critical_value:
    print("\nDecision: Reject the null hypothesis (H0). There is a significant association between the type of smart home device purchased and the customer satisfaction level.")
else:
    print("\nDecision: Fail to reject the null hypothesis (H0). There is no significant association between the type of smart home device purchased and the customer satisfaction level.")
```

Based on the Chi-Square statistic and the critical value, we can conclude whether there is a significant association between the type of smart home device purchased and the customer satisfaction level.
